```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load our beautifully preprocessed data
df = pd.read_csv('../datasets/processed/f1_analytics_master.csv')

# Convert Points to numeric just in case there are string artifacts
df['Points'] = pd.to_numeric(df['Points'], errors='coerce').fillna(0)

# Select the columns we want to compare
cols_to_compare = ['Points', 'Rating', 'Experience', 'Racecraft', 'Awareness', 'Pace', 'Starting Grid']

# Calculate the correlation matrix
corr_matrix = df[cols_to_compare].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation: Real World Points vs Video Game Stats", fontsize=16)
plt.show()
```
